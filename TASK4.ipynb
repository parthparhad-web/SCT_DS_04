```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\lenovo\Python_ds\RTA Dataset.csv")
df.head()
```

```python
print(df.head())
print(df.shape)
print(df.columns)
```

```python
df.info()
df.describe(include='all')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.fillna('Unknown', inplace=True)
```

```python
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M', errors='coerce')
```

```python
df['Hour'] = df['Time'].dt.hour
```

```python
def time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df['Time_of_Day'] = df['Hour'].apply(time_of_day)
```

```python
tod_counts = df['Time_of_Day'].value_counts()

plt.figure()
tod_counts.plot(kind='bar')
plt.xlabel("Time of Day")
plt.ylabel("Number of Accidents")
plt.title("Accidents by Time of Day")
plt.show()
```

```python
road_counts = df['Road_surface_type'].value_counts()

plt.figure()
road_counts.plot(kind='bar')
plt.xlabel("Road Surface Type")
plt.ylabel("Accident Count")
plt.title("Accidents by Road Surface Condition")
plt.show()
```

```python
weather_counts = df['Weather_conditions'].value_counts()

plt.figure()
weather_counts.plot(kind='bar')
plt.xlabel("Weather Condition")
plt.ylabel("Accident Count")
plt.title("Accidents by Weather Condition")
plt.show()
```

```python
severity_counts = df['Accident_severity'].value_counts()

plt.figure()
severity_counts.plot(kind='bar')
plt.xlabel("Accident Severity")
plt.ylabel("Count")
plt.title("Accident Severity Distribution")
plt.show()
```

```python
cause_counts = df['Cause_of_accident'].value_counts().head(10)

plt.figure()
cause_counts.plot(kind='bar')
plt.xlabel("Cause of Accident")
plt.ylabel("Count")
plt.title("Top Causes of Road Traffic Accidents")
plt.show()
```

```python
area_counts = df['Area_accident_occured'].value_counts()
plt.figure()
area_counts.plot(kind='bar')
plt.xlabel("Area Where Accident Occurred")
plt.ylabel("Number of Accidents")
plt.title("Accident Hotspots by Area Type")
plt.show()
```

```python
factors = [
    'Weather_conditions',
    'Road_surface_type',
    'Light_conditions',
    'Type_of_collision'
]
for col in factors:
    plt.figure()
    df[col].value_counts().plot(kind='bar')
    plt.xlabel(col.replace('_', ' '))
    plt.ylabel("Accident Count")
    plt.title(f"Accidents by {col.replace('_', ' ')}")
    plt.show()
```
